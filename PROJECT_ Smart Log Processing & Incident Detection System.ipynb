## 1️⃣ Clean Server Name

````python
def clean_server_name(name):
    """
    Task:
    - Remove leading/trailing spaces
    - Convert name to uppercase

    Example:
    " server-1 " → "SERVER-1"
    """
    # TODO: Implement cleaning logic
    return name.upper().strip()

print(clean_server_name('server-1  '))

````

## 2️⃣ Normalize Log Level

````python
def normalize_log_level(level):
    """
    Task:
    - Convert log level to uppercase
    - Allow only: ERROR, WARNING, INFO
    - Raise ValueError for invalid levels
    """
    # TODO: Implement validation
    upper_level = level.upper()
    allowed_levels = ['ERROR', 'WARNING', 'INFO']
    if upper_level not in allowed_levels:
        raise ValueError(f"Invalid log level: {level}. Allowed levels are {', '.join(allowed_levels)}.")
    return upper_level

print(normalize_log_level('Error'))
````

## 3️⃣ Validate Log Message

````python
def validate_message(message):
    """
    Task:
    - Ensure message is not None
    - Ensure message is a string
    - Raise TypeError or ValueError if invalid
    """
    # TODO: Implement validation
    if message is None:
        raise TypeError("Message cannot be None.")
    if not isinstance(message, (str)):
        raise TypeError("Message must be a string.")

    return message

print(validate_message('hello'))
````


## 4️⃣ Validate Usage Percentage


````python
def validate_usage(usage):
    """
    Task:
    - Must be int or float
    - Must be between 0 and 100
    - Raise appropriate error if invalid
    """
    # TODO: Implement validation
    if not isinstance(usage, (int, float)):
        raise TypeError("Usage must be a number.")
    if usage < 0 or usage > 100:
        raise ValueError("Usage must be between 0 and 100.")
    return usage

print(validate_usage(90))
````

## 5️⃣ Classify Severity

````python
def classify_severity(level, usage):
    """
    Severity Rules:

    ERROR   + usage >= 90  → CRITICAL
    ERROR   + usage >= 75  → HIGH
    WARNING + usage >= 80  → MEDIUM
    INFO                    → LOW
    """
    # TODO: Implement classification logic
    if level == 'ERROR':
        if usage >= 90:
            return 'CRITICAL'
        elif usage >= 75:
            return 'HIGH'
        else: # ERROR and usage < 75
            return 'LOW' # Default for lower severity errors
    elif level == 'WARNING':
        if usage >= 80:
            return 'MEDIUM'
        else: # WARNING and usage < 80
            return 'LOW' # Default for lower severity warnings
    elif level == 'INFO':
        return 'LOW'

print(classify_severity('ERROR', 89))
````

## 6️⃣ Detect Incident

````python
def detect_incident(severity):
    """
    Task:
    - Return True if severity is not LOW
    - Else return False
    """
    # TODO: Implement logic
    if severity != 'LOW':
        return True
    else:
        return False


print(detect_incident('CRITICAL'))
````

## 7️⃣ Process a Single Log Entry

````python
def process_single_log(raw_log):
    """
    Task:
    - Clean server name
    - Normalize log level
    - Validate message
    - Validate usage
    - Classify severity
    - Detect incident
    - Return structured dictionary

    Error Handling:
    - Catch and handle all possible exceptions
    - Return None for invalid logs
    - Always print 'Log Processed' using finally
    """
    try:
        # TODO: Extract raw values
        # TODO: Call helper functions
        # TODO: Build result dictionary
        # TODO: Return result
        server = clean_server_name(raw_log[0])
        level = normalize_log_level(raw_log[1])
        message = validate_message(raw_log[2])
        usage = validate_usage(raw_log[3])
        severity = classify_severity(level, usage)
        incident = detect_incident(severity)
        result = {
            "server": server,
            "level": level,
            "message": message,
            "usage": usage,
            "severity": severity,
            "incident": incident
        }
        return result

    except Exception as e:
        # TODO: Handle error safely
        print(f"Error processing log '{raw_log}': {e}")
        return None

    finally:
        print("Log Processed")
````

## 8️⃣ Process All Logs

````python
def process_all_logs(raw_logs):
    """
    Task:
    - Loop over all raw logs
    - Process each log using process_single_log
    - Ignore invalid logs
    - Return list of valid processed logs
    """
    processed_logs = []

    # TODO: Loop and collect valid logs
    for raw_log in raw_logs:
        processed_log = process_single_log(raw_log)
        if processed_log is not None:
            processed_logs.append(processed_log)
    return processed_logs
````

## 9️⃣ Generate Incident Summary


````python
def incident_summary(logs):
    """
    Task:
    - Count total logs
    - Count total incidents
    - Count severity-wise incidents

    Return format:
    {
        "total_logs": X,
        "incidents": Y,
        "severity_count": {
            "CRITICAL": A,
            "HIGH": B,
            "MEDIUM": C,
            "LOW": D
        }
    }
    """
    # TODO: Implement summary logic
    total_logs = len(logs)
    total_incident = sum(log['incident'] for log in logs if 'incident' in log)
    severity_count = {
        "CRITICAL": sum(1 for log in logs
                        if 'incident' in log and log['severity'] == 'CRITICAL'),
        "HIGH": sum(1 for log in logs
                    if 'incident' in log and log['severity'] == 'HIGH'),
        "MEDIUM": sum(1 for log in logs
                    if 'incident' in log and log['severity'] == 'MEDIUM'),
        "LOW": sum(1 for log in logs if 'incident' in log and log['severity'] == 'LOW')

    }
    return {
        "total_logs": total_logs,
        "incidents": total_incident,
        "severity_count": severity_count
    }
    print(incident_summary(logs))
````
